**Model Selection**

Train_test_split:Evaluate the performance of a model on unseen data. It involves dividing dataset into training set and testing set.

```python
#70%-Training data & 30%-Testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
print('Slope is',list(zip(x,model.coef_)))
```

```python
print("Constant is",model.intercept_)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
print('MAE is',mean_absolute_error(y_test,y_pred))
#MAPE
from sklearn.metrics import mean_absolute_percentage_error
print('MAPE is',mean_absolute_percentage_error(y_test,y_pred))
#MSE
from sklearn.metrics import mean_squared_error
print('MSE is',mean_squared_error(y_test,y_pred))
print('RMSE is',np.sqrt(mean_squared_error(y_test,y_pred)))
#R2_Score
from sklearn.metrics import r2_score
print('R2_score is',r2_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
#Creating object
dec=DecisionTreeRegressor()
dec.fit(x_train,y_train)
#Predicting using Test data
y_pred=dec.predict(x_test)
#Performance Evaluation
#--------------------------
#Mean Absolute Error(MAE)
from sklearn.metrics import mean_absolute_error
print('Decision tree MAE',mean_absolute_error(y_test,y_pred))
#Mean Absolute Percentage Error(MAPE)
from sklearn.metrics import mean_absolute_percentage_error
print('Decision tree MAPE',mean_absolute_percentage_error(y_test,y_pred))
#Mean Squared Error(MSE)
from sklearn.metrics import mean_squared_error
print('Decision tree MSE ',mean_squared_error(y_test,y_pred))
#Root Mean Squared Error(RMSE)
print('Decision tree RMSE',np.sqrt(mean_squared_error(y_test,y_pred)))
#R2_score
from sklearn.metrics import r2_score
print('Decision tree R2_score',r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
#Creating object
rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
#Predicting using Test data
y_predict=rf.predict(x_test)

#Performance Evaluation
#------------------
#Mean Absolute Error(MAE)
from sklearn.metrics import mean_absolute_error
print('Random Forest MAE',mean_absolute_error(y_test,y_predict))
#Mean Absolute Percentage Error(MAPE)
from sklearn.metrics import mean_absolute_percentage_error
print('Random Forest MAPE',mean_absolute_percentage_error(y_test,y_predict))
#Mean Squared Error(MSE)
from sklearn.metrics import mean_squared_error
print('Random Forest MSE ',mean_squared_error(y_test,y_predict))
#Root Mean Squared Error(RMSE)
print('Random Forest RMSE',np.sqrt(mean_squared_error(y_test,y_predict)))
#R2_score
from sklearn.metrics import r2_score
print('Random Forest R2_score ',r2_score(y_test,y_predict))
```
